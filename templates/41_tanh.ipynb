Implement the **tanh** activation function, its **backward pass**, and a **logit soft-capping** function.

### Part 1 — Forward

$$\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python

```

### Part 2 — Backward

The derivative of tanh has an elegant property — it can be expressed in terms of its own output:

$$\frac{d}{dx}\text{tanh}(x) = 1 - \text{tanh}^2(x)$$

```python
def tanh_backward(grad_output: torch.Tensor, tanh_output: torch.Tensor) -> torch.Tensor: ...
```

### Part 3 — Soft-Capping (Gemma 2)

Modern models like Gemma 2 use tanh to **soft-cap** logits, smoothly bounding them to $(-\text{cap}, +\text{cap})$:

$$\text{soft\_cap}(x) = \text{cap} \cdot \text{tanh}\left(\frac{x}{\text{cap}}\right)$$

```python
def soft_cap_logits(logits: torch.Tensor, cap: float) -> torch.Tensor: ...
```

### Rules
- Do **NOT** use `torch.tanh`, `F.tanh`, `torch.nn.Tanh`, or any built-in tanh
- Must support autograd (gradients should flow through `my_tanh`)
- `tanh_backward` should be a **manual** computation, not using autograd
- `soft_cap_logits` should use your `my_tanh`
